# Aim:
To implement and visualize DBSCAN clustering on synthetic datasets and the Pima Indians Diabetes dataset, including optimal parameter tuning, cluster detection, and graphical representation of dense regions and outliers.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_blobs
dataset,classes = make_blobs(n_samples=250,n_features=2, centers=1,cluster_std=0.3, random_state=1)
```

```python
plt.scatter(dataset[:,0],dataset[:,1])
```

```python
classes
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan=DBSCAN(eps =0.3,min_samples =20)
```

```python
pred = dbscan.fit_predict(dataset)
```

```python
pred
```

```python
plt.scatter(dataset[:,0],dataset[:,1],c=pred)
```

```python
outlier_index = np.where(pred==-1)
outlier_val = dataset[outlier_index]
```

```python
plt.scatter(dataset[:,0],dataset[:,1])
plt.scatter(outlier_val[:,0],outlier_val[:,1])
```

```python
dbscan=DBSCAN(eps =0.2,min_samples =20)
pred = dbscan.fit_predict(dataset)
plt.scatter(dataset[:,0],dataset[:,1],c=pred)
```

```python
dbscan=DBSCAN(eps =0.3,min_samples =10)
pred = dbscan.fit_predict(dataset)
plt.scatter(dataset[:,0],dataset[:,1],c=pred)
```
